```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('forestfires_original.csv')
```

```python
data.head()
```

```python
data = data.drop(['X'], axis = 1)
data = data.drop(['Y'], axis = 1)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data.corr(),annot=True,cmap='viridis',linewidths=.5)
```

```python
data['area']
```

```python
from collections import Counter
```

```python
x, y = [], []
for i, j in Counter(sorted(data['area'])).items():
  x.append(i)
  y.append(j)
plt.plot(x, y)
print(x)
```

```python
bins = [-1, 0, 20, 100, 1000, np.inf]
names = ['No threat', 'Light', 'Moderate', 'Severe', 'Extremely severe']
```

```python
pd.cut(data['area'], bins, labels=names)
```
